## Esercizio 1

**Testo**

Si desidera stimare il tempo medio di CPU usato agli utenti di un server in modo da affermare con il 95% di confidenza il che il valore stimato non disti più di 0.5 secondi dal valore vero.

Il tempo di CPU è distribuito secondo una normale con $\sigma^2 = 2.25$.

Quale deve essere la dimensione $n$ del campione?

**Svolgimento**

* Il valore stimato è $\bar{X}$
* Il livello di significatività è pari a $0.05$
    * Infatti $1 - \alpha = 0.95 \to \alpha = 0.05$
* La popolazione è normalmente distribuita con varianza nota pari a $\sigma^2 = 2.25$

L'intervallo che cerchiamo è simmetrico rispetto alla stima puntuale $\bar{X}$ del valore vero del parametro $\mu$ e desideriamo che esso sia tale per cui il valore $\mu$ ne sia contenuto con probabilità 0.95.

Formalmente

$$P(\mu \in (\bar{X} - z_{1 - \frac{\alpha}{2}} \frac{\sigma}{\sqrt{n}}, \bar{X} + z_{1 - \frac{\alpha}{2}} \frac{\sigma}{\sqrt{n}}) ) = 0.95$$

Per risolvere l'esercizio dobbiamo trovare $n$ tale che 

$$0.5 = z_{1 - \frac{\alpha}{2}} \frac{\sigma{}}{\sqrt{n}} $$

**Nota bene**: $z_{1 - \frac{\alpha}{2}} \frac{\sigma{}}{\sqrt{n}} $ pone il vincolo su quanto il valore stimato possa differirsi dal valore vero.

Quindi calcoliamo $z_{1 - \frac{\alpha}{2}} \frac{\sigma{}}{\sqrt{n}}$. 

$$z_{1 - \frac{0.05}{2}} \frac{1.5}{\sqrt{n}} = z_{0.975} \frac{1.5}{\sqrt{n}}$$

Troviamo inanzitutto il valore di $Z_{0.975}$ guardando sulle tavole: 1.96

Quindi $1.96 \cdot \frac{1.5}{\sqrt{n}} = 0.5$

Facendo le opportune semplificazioni $n = 35$

## Esercizio 2

**Testo**

Un segnale $\mu$ viene trasmetto da $A$ a $B$. Il segnale ricevuto in $B$ è distribuito normalmente con media $\mu$ e varianza $\sigma^2$ incognita. Se vengono trasmessi 9 segnali successivi con valore $\mu$ uguale e ricevuti in B i seguenti valori: $5 - 8.5 - 12 - 15 - 7 - 9 - 7.5 - 6.5 -  10.5$. Si calcoli un intervallo di confidenza con il 95% di confidenza per $\mu$.

**Svolgimento**

Siamo nel caso del *calcolo dell'intervallo di confidenza per la media con popolazione normalmente distribuita e varianza non nota*. 

Quindi l'intervallo di convidenza si presenta in una forma:

$$\mu \in (\bar{X} - t^{n-1}_{1 - \frac{\alpha}{2}} \frac{S^\wedge}{\sqrt{n}}, \bar{X} + t^{n-1}_{1 - \frac{\alpha}{2}} \frac{S^\wedge}{\sqrt{n}})$$

Iniziamo con il calcolo di $\bar{X}$, media campionaria.

$$\frac{5+8.5+12+15+7+9+7.5+6.5+10.5}{9} = 9$$

Inoltre possiamo calcolare la varianza campionaria corretta $S^{\wedge^2}$

$$S^{\wedge^2} = \frac{1}{n-1} \cdot \sum{X_i - X}^2 = 9.5$$

Dobbiamo anche calcoalre $t^{n-1}_{1 - \frac{\alpha}{2}}$

Abbiamo $\alpha = 0.05$, conosciamo $n$ ora ci basta cercare sulle tavole della t di Student il valore $t^8_{0.975} \to 2.3$

Abbiamo quindi tutti i dati per definire l'intervallo:

$$\mu \in (9 - 2.3 \frac{3.08}{3}, 9 + 2.3 \frac{3.08}{3})$$

## Esercizio 4

**Testo**

Una fabbrica produce prodotti la cui durata di vita si distribuisce con una varianza di $\sigma^2 = 0.25$. Si supponga di estrarre un campione casuale di 49 pezzi aventi durata media pari a 5.5 anni. Costruire un intervallo di confidenza al 95% per la durata media dei prodotti.

**Svolgimento**

La variabile $\bar{X}$ relativa alla vita media è distribuita secondo una normale non standard. 

$$\bar{X} \sim N(\mu, \frac{\alpha}{\sqrt{n}})$$

Normalizziamo e troviamo:

$$\frac{\bar{X} - \mu}{\frac{\alpha}{\sqrt{n}}} = Z \in [-z_{1-\frac{\alpha}{2}}, +z_{1 - \frac{\alpha}{2}}]$$

che significa

$$P(Z \in [-z_{1-\frac{\alpha}{2}}, +z_{1 - \frac{\alpha}{2}}]) = 1 - \alpha$$

cioè 

$$P(-z_{1 - \frac{\alpha}{2}} \leq \frac{\bar{X} - \mu}{\frac{\alpha}{\sqrt{n}}} \leq +z_{1 - \frac{\alpha}{2}}) = 1 - \alpha$$

Esplicitando l'intervalo in funzione di $\mu$:

$$P(\bar{X} - z_{1 - \frac{\alpha}{2}} \frac{\alpha}{\sqrt{n}} \leq \mu \leq \bar{X} + z_{1 - \frac{\alpha}{2}} \frac{\alpha}{\sqrt{n}}) = 1 - \alpha$$

Che sono tutti valori noti o facilmente calcolabili.

**Svolgimento**


Sia $X$ la variabile casuale che denota la frequenza del segnale radio emesso.

Osserviamo che è distribuita normalmente con varianza nota ($\sigma^2 = 900$).

Calcoliamo $\bar{x} = 610.27$. Questa è una stima di $\mu$.

Quini $X \sim N(610.27, 900)$.

Ora dobbiamo cercare $P(590 \leq X \leq 610)$

Dato che $X$ non è distribuita secondo una normale standard, dobbiamo effettuare la standardizzazione.

Quindi $P(-0.68 \leq z \leq -0.01) = P(0.01 \leq 0.68)$.

## Esercizio 7

Si prendono 100 pasticcini. Il loro peso medio è 35. Lo sqm di tutti i pasticcini prodotti è pari a 4.

* Si trovi l'intervallo di confidenza per il peso medio di tutti i pasticcini prodotti a livello di confidenza del 98%
* Di quanto deve aumentare la numerosità campionaria se si vuole che l'ampiezza dell'intervallo si dimezzi?
* Si determini quanti pasticcini occorre estrarre se si vuole che lo stimatore del peso si discosti dal vero peso medio di 1 grammo con probabilità del 96%

**Svolgimento**

Visto che sono più di 100 pasticcini possiamo assimilare la va $X$ a $N(\mu, \frac{\sigma^2}{n})$ con $\mu$ ignota e $\sigma=4$.

L'intervallo di confidenza per la media $\mu$ con livello di convidenza $1 - \alpha$

$$P(\bar{X_n} - z_{1 - \frac{\alpha}{2}} \frac{\sigma}{\sqrt{n}} \leq \mu \leq \bar{X_n} + z_{1 - \frac{\alpha}{2}} \frac{\sigma}{\sqrt{n}}) = 1 - \alpha$$

che nel nostro caso è

$$P(35 - z_{1 - 0.02/2} \frac{4}{\sqrt{100}} \leq \mu \leq 35 + z_{1 - 0.02/2} \frac{4}{\sqrt{100}}) = 0.98$$
