```r
# Esercizio 1
# Quattro monete bilanciate vengono lanciate. 
# Assumendo l’indipendenza dei risultati, qual è la probabilità di ottenere due testa e due croce?
nLanci = 4
nEsperimentiCorretti = 2
p = 1/2

dbinom(nEsperimentiCorretti, nLanci, p)

plot(c(0:nLanci), dbinom(0:nLanci, nLanci, p=p), type="h")
lines(c(0:nLanci), dbinom(0:nLanci, nLanci, p=p), col="red")
text(c(0:nLanci), dbinom(0:nLanci, nLanci, p=p), dbinom(0:nLanci, nLanci, p=p))
```

```r
# Esercizio 2
# E' noto che gli item prodotti saranno difettosi con p=0.1.
# Quale è la probabilità che su 3 items al più uno è difettoso?
nItems = 3
nMaxDifettosi = 1
pDifettoso = 0.1

pbinom(nMaxDifettosi, nItems, pDifettoso)

# equivalente al fare
partial = 0
for(i in 0:nMaxDifettosi) {
    partial = partial + dbinom(i, nItems, pDifettoso)
}
partial
```

```r
# Esercizio 3
pDifettoso = 0.01
nDischiPerConfezione = 10
nDischiRottiPerRimborsare = 1
# Quale è la probabilità che una confezione venga rimborsata

1 - pbinom(nDischiRottiPerRimborsare, nDischiPerConfezione, pDifettoso)

# equivalente a 
pbinom(nDischiRottiPerRimborsare, nDischiPerConfezione, pDifettoso, lower.t=F)

# equivalente a 
partial = 0
for(i in 0:nDischiRottiPerRimborsare) {
    partial = partial + dbinom(i, nDischiPerConfezione, pDifettoso)
}
1 - partial
```

```r
nScatoleAcquistate = 3
nScatoleRestituite = 1
dbinom(
    nScatoleRestituite, 
    nScatoleAcquistate, 
    pbinom(nDischiRottiPerRimborsare, nDischiPerConfezione, pDifettoso, lower.t=F)
)
```

```r
# Esercizio 4

lambdaErrori = 1
nErroriMax = 0

ppois(nErroriMax, lambda=lambdaErrori, lower.t=FALSE) 
# probabilità che ci sia più di nErroriMax

# equivale a 
1 - ppois(nErroriMax, lambda=lambdaErrori)
# 1 - probabilità che ci sia fino a nErroriMax
```

```r
# lambda = 1
# probabilità compresa tra 3 e 5

sum(dpois(3:5, 1))
```

```r
# grafico della distribuzione di poisson lambda=1

plot(0:5, dpois(0:5, 1), type="h", xlab="X")
```

```r
# un calciatore ha probabilità di segnare pari a p
# fare un tiro è un evento bernoulliano
p = 0.812

# probabilità che sbagli 5 tiri prima di segnarne uno
dgeom(5, p=p)

# probabilità che sbagli al massimo 5 tiri prima di segnarne uno
pgeom(5, p=p)

# probabilità che sbagi almeno 5 calci di rigore prima di segnarne uno
pgeom(4, p=p, lower.tail = FALSE)
1 - pgeom(4, p=p)
```
