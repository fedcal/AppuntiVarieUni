# Distribuzioni Notevoli

Si parla di **distribuzione** di una variabile intendendo indifferentemente la sua ripartizione o la sua densità.

Per indicare che una variabile aleatoria è distribuita secondo una certa distribuzione si usa la notazione $$X \sim F$$

## Distribuzioni Bernoulliana

Una variabile aleatoria $X$ è detta distribuita secondo una Bernoulliana di parametro p (indicata con $X \sim B(p)$) se essa può assumere solo i valori 1 e 0, rispettivamente con probabilità $p$ e $1-p$

Questo si traduce in 
$$p_X(t) = \begin{cases}
1-p & \text{se } t=0 \\
p & \text{se } t=1 \\
0 & \text{altrimenti}
\end{cases}$$

$$F_X(t) = \begin{cases}
0 & \text{se } t < 0 \\
1-p & \text{se } 0 < t < 1 \\
1 & \text{se } t \geq 1
\end{cases}$$

Sono quindi variabili di Bernoulli tutte quelle che individuano il verificarsi di uno specifico evento e che valgono 1 se questo si verifica e 0 altrimenti.

### Media e Varianza

$$E[X] = 0 \cdot (1-p) + 1 \cdot p$$

$$V[X] = [0^2 \cdot (1-p) + 1^2 \cdot p] - p^2 = (1-p)p$$

## Distribuzione Binomiale

Siano $X_1$, $X_2$, $X_n$, $n$ variabili bernoulliane di identico parametro $p$ e stocasticamente indipendenti. Sia $X$ una variabile aleatoria definita come somma delle $X_i$

Una tale variabile aleatoria è detta distribuita secondo una **binomiale con parametri n,p** $$X \sim Bin(n,p)$$

Tale variabile può assumere qualsiasi valore intero $k$ compreso tra 0 ed $n$ in accordo alla seguente probabilità

$$P(X=k) = \binom{n}{k} p^k (1-p)^{n-k}$$

* $p^k (1-p)^{n-k}$ fornisce le probabilità che $k$ delle $n$ variabili $X_i$ assumano valore di 1 e che le restanti $n-k$ assumano valore 0.
* $\binom{n}{k}$ invece esprime il numero di combinazioni possibili per cui $k$ variabili valgono $1$ e $n-k$ valgono 0.

$$
p_X(t) = \begin{cases}
\binom{n}{t} p^t (1-p)^{n-t} & t \in \{0,1,\ldots,n\} \\
0 & \text{altrimenti}
\end{cases}
$$

$$
F_X(t) = \sum_{0 \leq k \leq n: k \leq t} \binom{n}{k}p^k (1-p)^{n-k}
$$

Una distribuzione binomiale viene usata nella definizione di variaibli che contano le realizzazioni di eventi, quando questi siano indipendenti e con identica probabilità di verificarsi.


```r
numeroVariabili = 20
p = 0.7
asseX = seq(0, numeroVariabili, by=1) # produce l'asse X
asseY = dbinom(asseX, numeroVariabili, p)
plot(asseX,asseY)
```

### Media e Varianza

Visto che le variabili sono indipendenti è facile calcoalare la media e la varianza

* $E[X] = E[X_1 + \ldots + X_n] = E[X_1] + \ldots + E[X_n] = n \cdot p$
* $V[X] = V[X_1 + \ldots + X_n] = V[X_1] + \ldots + V[X_n] = n \cdot (1-p) \cdot p$

## Distribuzione di Poisson

Caso particolare della distribuzione binomiale. Si ottiene quando il numero di variabili $X_i$ tende ad essere infinio mentre il valore di $p$ tende a zero in modo tale che il prodotto $n \cdot p$ resti costante.

Definiamo in questo caso $\lambda = n \cdot p$ e quindi diremmo che $X ~ Poi(\lambda)$

$$P(X=k) = \frac{\lambda^k}{k!} e^{-\lambda}$$

$$p_X(t) = \begin{cases}
\frac{\lambda^t}{t!} e^{-\lambda} & t \in \{0,1,\ldots\} \\
0 & \text{altrimenti}
\end{cases}$$

$$F_X(t) = \sum_{k \leq t} \frac{\lambda^k}{k!} e^{-\lambda}$$

La distribuzione di Poisson viene utilizzata quando si considerino grandi popolazioni di individui in cui ogni individuo ha una probabilità $p$, molto molto piccola, di essere soggetto ad un possibile evento in esame. E' anche detta distribuzione degli eventi rari.

#### Media e Varianza

* $E[X] = n \cdot p = \lambda$
* $V[X] = n (1-p) p = np - np^2 = \lambda - \frac{\lambda^2}{2}$
    * Questa quantità diviene uguale a $\lambda$ quando $n$ tende ad infinito per cui 
    * $V[X] = \lambda$

```r
## probabilità che si verifichi asseX dato che la media è lambda
lambda = 7
asseX = seq(0, 20, by=1)
asseY = dpois(asseX, lambda)
plot(asseX,asseY)
```

## Distribuzione geometrica

$X \sim Geo(p)$ se può assumere qualsiasi valore intero non negativo $k$ con probabilità $$P(X = k) = p(1-p)^k$$.

$$p_X(t) = \begin{cases}
p(1-p)^t & t \in N \\
0 & \text{altrimenti}
\end{cases}$$

$$F_X(t) = \sum_{k \leq t} p(1-p)^k$$

### Esempio

Si consideri un'esperimento di verifica funzionamento di una macchina. $p$ è la probabilità che ad un certo istante la macchina si guasti. $p_X(t)$ è la probabilità che la macchina si guasti al tempo $t+1$ (fino all'istante $t$ funziona)

### Media e Varianza

* $E[X] = \frac{1-p}{p}$
* $V[X] = \frac{1-p}{p^2}$

```r
p = 0.3 # probabilità che qualcosa non si verifica
asseX = seq(0, 10, by=1)
asseY = dgeom(asseX, p)
plot(asseX,asseY)
```

### Assenza di memoria

La distribuzione geometrica gode della probabilità di assenza di memoria.

$$P(X = k + m | X \geq m) = P(X=k)$$ 

Per capire, supponiamo X uguale al tempo di vita di una macchina soggetta a guasti. E supponiamo di aver rilevato che per $m$ unità di tempo essa non si sia guastata. La probabilità che la macchina si guasti nell'istante $k+m$ è uguale alla probabilità che la macchina si guasti all'istante $k$. 

La proprietà di assenza di memoria asserisce che il tempo trascorso da quando abbiamo inziato ad esaminare il funzionamnento della macchina non influisce sulla distribuzione del tempo restante al verificarsi al guasto.

## Distribuzione uniforme

Rappresenta la più semplice distribuzione continua e viene adottata nel caso in cui la variabile condierata possa assumere qualsiasi valore compreso in un dato intervallo con probabilità costante.

$$X \sim [a,b]$$

$$f_X(y) = \begin{cases}
\frac{1}{b-a} & t \in [a,b] \\
0 & \text{ altrimenti}
\end{cases}$$

$$
F_X(t) = \begin{cases}
0 & t < a \\
\frac{t-a}{b-a} & t \in [a,b] \\
1 & t > b
\end{cases}
$$

Descrive bene situazioni dove le variabili possono assumere valori in intervalli finiti di R con probabilità uniforme ovverto tale da essere identica per intervalli di ampiezza medesima.

### Valore atteso e varianza

* $E[X] = \frac{a+b}{2}$
* $V[X] = \frac{(b-a)^2}{12}$

## Distribuzione triangolare

Assegna alla densità di probabilità valori maggiori al centro del supporto e minori in prossimità degli estremi.

$$X \sim T[a,b]$$

$$
f_X(t) = \begin{cases}
4 \cdot \frac{(t-a)}{(b-a)^2} & t \in [a, \frac{a+b}{2}) \\ \\
4 \cdot \frac{(b-t)}{(b-a)^2} & t \in [\frac{a+b}{2}, b] \\ \\
0 & \text{altrimenti}
\end{cases}
$$

$$
F_X(t) = \begin{cases}
0 & t < a \\ \\
2 \cdot \frac{(t-a)^2}{(b-a)^2} & t \in [a, (a+b)/2) \\ \\
1 - 2 \cdot \frac{(b-t)^2}{(b-a)^2} & t \in [\frac{a+b}{2}, b] \\ \\
1 & t > b
\end{cases}
$$

### Valore atteso e varianza

* $E[X] = \frac{a+b}{2}$
* $V[X] = \frac{(b-a)^2}{24}$


## Distribuzione Esponenziale

Utile nello studio delle variabili che descrivono i tempi occorrenti al verificarsi di un evento (ad esempio tempo di attesa di una vendita o di una costruzione di un immobile).

$$X \sim Exp(\lambda)$$

$$
f_X(t) = \begin{cases}
\lambda \cdot e^{- \lambda t} & t \geq 0\\
0 & t < 0
\end{cases}
$$

$$
F_X(t) = \begin{cases}
1 - e^{- \lambda t} & t \geq 0 \\
0 & t < 0
\end{cases}
$$

Questa distribuzione gode della proprietà di assenza di memoria. 

$$P(X > s + t \, | \, X > s) = P(X > t)$$
### Valore Atteso e Varianza

* $E[X] = \frac{1}{\lambda}$
* $V[X] = \frac{1}{\lambda^2}$

## Distribuzione Normale

$$X \sim N(\mu, \sigma)$$

$$f_X(t) = \frac{1}{\sqrt{2 \cdot \pi \cdot \sigma^2}} \cdot e^{- \frac{(t-\mu)^2}{2 \cdot \sigma^2}}$$ 

Questa distribuzione è fondamentale per il teorema centrale del limite.  

La variabile aleatoria ha supporto su tutto l'asse reale.

La funzione di densità ha un punto di flesso in $\mu - \sigma$ e $\mu + \sigma$, la moda coincide con la media.

### Valore atteso e varianza

* $E[X] = \mu$
* $V[X] = \sigma^2$



```r
asseX = seq(-5, 5, by=0.1)
asseY = dnorm(asseX)
plot(asseX,asseY, type="l")
```

### Calcolo della P(X)

Data una variabile aleatoria distribuita secondo una normale non è semplice calcolare $P( X \in [a,b])$, perchè vorrebbe dire calcolare l'integrale tra $a,b$ di $f_X(t)$.

Per questa ragione esistono delle tabole che si riferiscono alla distribuzione normale standard, ovvero con parametri $\mu = 0$ e $\sigma = 1$ che forniscono i valori di $$\int^{z}_{0} f_X(t)$$ per molti valori di $z \in R_+$.

Quando vogliamo determinare le probabilità associata ad una generica $X \sim N(\mu, \sigma)$ è dobbiamo **normalizzare** facendo $$Z = \frac{X - \mu}{\sigma}$$.

Infatti se $X \sim N(\mu,\sigma)$ allora $Z \sim N(0,1)$.

$$P(X \in [a,b]) = P(a \leq X \leq b) = P(Z \in [\frac{a - \mu}{\sigma}, \frac{b- \mu}{\sigma}])$$

### Esempio

$\mu = 10$ e $\sigma = 1$, determinare $P(X \in [9.2, 11.35])$

$$P(Z \in [ \frac{9.2 - 10}{1}, \frac{11.35-10}{1} ]) = P(Z \in [-0.8, 1.35])$$

La normale è simmetrica per cui $P(Z \in [-0.8, 0]) = P(Z \in [0, 0.8])$.

Inoltre dalla tavola $P(Z \in [0, 0.8]) = 0.2881$ e $P(Z \in [0, 1.35]) = 0.4115$.

Sommandole $P( Z \in [-0.8, 1.35]) = 0.6996$

### Regole di calcolo per normali

* $P(Z < n) = P(Z \geq -n) = 1 - P(Z \geq n)$

Per integrali nella forma $$\int^b_{- \infty} f(u) du$$

* Se $b > 0 = \frac{1}{2} + \int^b_0 f(u) du$
* Se $b < 0 = \frac{1}{2} - \int^{-b}_0 f(u) du$

Inoltre $$\int^{\infty}_{a} f(u)du = 1 - \int^a_{- \infty} f(u)du$$

Ed anche 

$$\int^b_a f(u) du = \int^b_{- \infty} f(u) du - \int^a_{- \infty} f(u) du$$

#### Altre proprità

* $P(X \in [\mu - \sigma, \mu + \sigma]) = 0.683$
* $P(X \in [\mu - 2 \sigma, \mu + 2 \sigma]) = 0.954$
* $P(X \in [\mu - 3 \sigma, \mu + 3 \sigma]) = 0.997$


**Chiusura rispetto all'operazione di somma di variabili stocasticamente indipendenti**

Se $X1 \sim N(\mu 1, \sigma 1), X2 \sim N(\mu 2, \sigma 2)$ sono indipendenti, allora $Y= X1 + X2$ è $$ Y \sim N(\mu 1 + \mu 2, \sqrt{\sigma 1 + \sigma 2})$$

## Altre distribuzioni

### Distribuzione chi-quadro

Siano date $n$ variabili con distribuzione normale di parametri 0 ed 1 (normali standardizzate) ed indipendenti tra loro. 

Sia $X$ la somma dei quadrat delle $n$ variabili aleatorie. $X$ è detta distribuita secondo *Chi-Quadro* con $n$ gradi di libertà. 

Assume sempre un valore non negativo.

### Distribuzione t di Student

Siano $Z \sim N(0,1)$ e $Y \sim \chi^2_n$ due variabili indipendenti. Sia $X$ definita come $$X = \frac{Z}{\sqrt{\frac{Y}{n}}}$$ Una tale variabile è definita secondo una *t di Student* con $n$ gradi di libertà $X \sim t_n$

### Distribuzione F

Siano $U \sim \chi^2_m$ e $V \sim \chi^2_n$ due variabili indipendenti. Sia poi $X$ una variabile aleatoria definita come $$X = \frac{\frac U m}{\frac V n}$$
Si dice che $X$ è distribuita secondo una F con $m,n$ gradi di libertà $$ X \sim F(m,n)$$
