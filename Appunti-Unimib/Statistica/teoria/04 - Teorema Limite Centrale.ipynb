# Teoremi di Convergenza

## Successioni Convergenti

Consideriamo una successione $\{ X_n, n \in N \}$ di variabili aleatorie. Sia $F_n$ la funzione di ripartizione della variabile $X_n$. 

Si dice che la **successione converge** in distribuzione alla variabile $X$ se vale $$\lim_{n \to \infty} F_n(t) = F(t)$$ per ogni $t \in R$ che sia punto di continuità per la F.

Si usano le notazioni $X_n \xrightarrow{d} X$ e $F_n \xrightarrow{d} F$. 

Questa convergenza è anche detta *convegenza di distribuzione*.

### Esempio

$$F_n(t) = \begin{cases}
0 & t \leq 0 \\
t^{\frac{n}{n+1}} & 0 < t < 1 \\
1 & y \geq 1
\end{cases}$$

Osserviamo che per ogni $t \in (0,1)$ vale che $$\lim_{n \to \infty} t^{\frac{n}{n+1}} = t$$

Qundi $X_n \xrightarrow{d} X$ con X che ha la seguente funzione di ripartizione.

$$F(t) = \begin{cases}
0 & t \leq 0 \\
t & 0 < t < 1 \\
1 & t \geq 1
\end{cases}
$$

**Nota bene**: la funzione di ripartizione generica potrebbe non ammettere limite! Di conseguenza è impossibile fare tutte queste considerazioni.

## Legge dei Grandi Numeri

Consideriamo una successione $$\{ X_i, i \in N_+ \}$$ di variabili aleatorie indipendenti ed identicamente distribuite (IID: cioè hanno tutte la stessa distribuzione).


Consideriamo poi
$$\bar{X_n} = \frac{X_1 + X_2 + \ldots + X_n}{n}$$
detta media aritmetica $n$-sima della successione.

Se le variabili $X_i$ hanno valore atteso $\mu$ e varianza $\sigma^2$ allora vale che
$$\bar{X_n} \xrightarrow{d} M$$
dove $M$ è una variabile aleatoria con $\mu = 1$

Tale proprietà ci permette di enunciare formalmente la **legge dei grandi numeri**.

### Definizione

Consideriamo una successione di variabili aleatorie IID per cui esistono valore atteso e varianza. Possiamo affermare che la successione 
$$\{ \bar{X_n}, n \in N_+ \}$$
delle corrispondenti medie aritmentiche tende, al crescere di $n_i$, ad una variabile che assume come valore atteso $\mu$

In altre parole se consideriamo una successione $\{ x_i, i \in N_+ \}$ di realizzazione delle variabili $\{ X_i, i \in N_+ \}$ e se consideriamo la successione 
$$\{ \bar{x_n}, n \in N_+ \}$$
delle corrispondenti realizzazioni delle medie aritmetiche, abbiamo che questa seconda successione tende, per $n \to \infty$ al valore
$$E[X_i] = \mu$$

## Teorema Limite Centrale

La legge dei Grandi Numeri assicura la convergenza $\bar{X_n} \xrightarrow{d} M$ ma non specifica nulla circa la rapidità con cui questa avviene. Non sappiamo infatti dire per quale valore di $n$ sarà lecito pensare che una realizzazione $\bar{x_n}$ assuma valore $\mu$ (o comunque un valore molto vicino a $\mu$)

E' intuitivo pensare che la convergenza sia tanto più rapida quanto la varianza sarà piccola. 

Quanto appena ipotizzato viene confermato e formalizzato tramite il **Teorema Limite centrale**. 

### Definizione

Sia $\{ X_i, i \in N_+ \}$ una successione di V.A. IID con stesso valore atteso $\mu$ e varianza $\sigma^2$ (esistenti e finiti).

Consideriamo la variabile
$$S_n = X_1 + \ldots + X_n $$
vale che
$$S_n \xrightarrow{d} X \sim N(n \cdot \mu, \sqrt{n} \cdot \sigma) = N(n \cdot \mu, n \cdot \sigma^2)$$
ovvere che $S_n$ converge in distribuzione ad una variabile distribuita come una normale di media $n \cdot \mu$ e deviazione standard $\sqrt{n} \cdot \sigma$

Tale proprietà è una forma debole del **teorema del limite centrale**.

Osserviamo inanzitutto che vale $\bar{X_n} = \frac{S_n}{n}$.

* Ricordando che $E[aX+b] = aE[X]+b$ allora:

$$E[\bar{X_n}] = E[\frac{S_n}{n}] = \frac{1}{n} E[S_n] = \frac{1}{n} E[X_1 + \ldots + X_n] = \frac{1}{n} (n \cdot \mu) = \mu$$

* Ricordiamo che vale $V[aX+b] = a^2 \cdot V[X]$ allora:

$$V[\bar{X_n}] = V[\frac{S_n}{n}] = \frac{1}{n^2} V[S_n] = \frac{1}{n^2} V[X_1 + \ldots + X_n] = \frac{1}{n^2} (n \cdot \sigma^2) = \frac{\sigma^2}{n}$$

Concludiamo infine che 

$$\bar{X_n} \xrightarrow{d} X \sim N(\mu, \frac{\sigma} {\sqrt{n}}) = N(\mu, \frac{\sigma^2}{n})$$

### In pratica

Il TLC ci dice che per $n$ sufficientemente grandi ($n \geq 30$) allora possiamo approssimare le bariabili $S_n$ ed $\bar{X_n}$ con delle variabili aventi distribuzione normale i cui parametri dipendendono dalle variabili $X_i$.
