# Regressione Lineare

Ci si chiede se tra m+1 vari caratteri di una popolazione puo' esistere legame di dipendenza funzionale cioe' se esiste una relazione del tipo 

$$Y = f(X_1, \ldots, X_m)$$

Tale espressione viene detta equazione di regressione lineare del carattere Y rispetto agli altri m caratteri.

E' detta variabile di risposta la $Y$ mentre sono dette variabili esplicative le $X_1, \ldots X_m$.


Con analisi di regressione si intende invece la verifica dell'ammissibilita' di un relazione $Y = f(X_1, \ldots, X_m)$

In questo capitolo ci limitiamo ad osservare solo equazioni di regressione nella quale si suppone la $f$ funzione lineare nella variabili $X_m$ e con un addendo aleatorie $E$.

$$Y = a_0 + a_1 \cdot X_1 + \ldots + a_m \cdot X_m + E$$

con $E \sim N(0, \sigma^2)$.

Se compare una sola variabile stiamo trattando la regressione lineare semplice, altrimenti quella multipla. 

E e' detto residuo o errore. 

Sia il residuo sia i valori delle varie $X_m$ cambiano per ogni individuo della popolazione quindi e' meglio definire 

$$Y_i = a_0 + a_1 \cdot X_{1,i} + \ldots + a_m \cdot X_{m,i} + E_i$$




## Stima delle costanti del modello

Consideriamo una popolazione bidimensionale ed assumiamo che tra i due caratteri esista un legame di dipendenza esprimibile tramite un'equazione di regressione lineare. 
Per ogni individuo si ha quindi:

$$\alpha_0 + \alpha_1 \cdot X_1 + E_i$$

In primo luogo ci interessa determinare il valore delle costanti $\alpha_0$ e $\alpha_1$.

Ovviamente possiamo solo fare stime riguardo a queste costanti. E' impossibile trovare il valore esatto visto che abbiamo a disposizione solo $N$ individui.

Per determinare le stime $a_0$ e $a_1$ dobbiamo fare tre ipotesi. 

* $E[E_1] = 0$
* $V[E_1] = \sigma^2$, ipotesi di omoschedasticita'
* Le variabili E_i sono incorrelate tra loro

Pur supponendo la validita' delle ipotesi non conosciamo il valore della varianza (che epr ora teniamo come costante nota).

Utilizziamo gli stimatori puntuali 

$$A_0 = \bar Y - A+1 \cdot \bar X$$

$$A_1 = \frac{\sum (X_i - \bar X) \cdot (Y_i - \bar Y)}{\sum (X_i - \bar X)^2}$$


> Tali stimatori sono i valori per cui risulta minima la quantita' $$\sum [A_1 \cdot X_i + A_0 \cdot Y_i]^2$$ 

Quindi 

$$a_0 = \bar y - a_1 \cdot \bar x$$


$$a_1 = \frac{\sum (x_i - \bar x) \cdot (y_i - \bar y)}{\sum (x_i - \bar x)^2}$$

![](images/8-01.png)

Da queste due formule notiamo che se la dispersione delle osservazioni e' piccola allora il denominatore e' molto piccolo e le varianze molto grandi (in questo caso quindi le probabilita' che le stime siano lontane dai valori reali cresce).


------

Se vogliamo effettuare delle stime intervallari per le costanti $\alpha_0$ e $\alpha_1$ dobbiamo aggiungere una nuova ipotesi. Cioè che le variabili $E_i$ siano normalmente distribuite.

In questo caso è possibile dimostrare che anche gli stimatori $A_0$ e $A_1$ siano normalmente distribuiti.

Rimane ancora da individuare il parametro $\sigma^2$, che possiamo sostituirlo con una stima 

$$S^2_{RES} = \frac{1}{n-2} \sum (Y_i - \bar Y_i^\wedge)^2$$

dove $Y_i^\wedge$ sono i valori teorici che dovrebbero essere assunti dal carattere $Y$ se si trovassero sulla retta di regressone stimata cioè

$$Y^\wedge_i = a_0 + a_1 \cdot X_i$$


$S^2_{RES}$ è la varianza campionaria  che descrive lo scostamento tra le osservazioni $y_i$ del carattere Y e le loro stime ($y^\wedge = a_0 + a_1 \cdot x_i$) ottenute tramite la regressione.

Inoltre $$E[S^2_{RES}] = \sigma^2$$

-----

Otteniamo gli intervalli di confidenza per $\alpha_0$ e $\alpha_1$ che risultano essere

$$[a_0 - t_{1- \alpha/2} \cdot \sqrt{s^2_{RES} \cdot [\frac{1}{n} + \frac{\bar x^2}{s^2_x}]}, a_0 + t_{1- \alpha/2} \cdot \sqrt{s^2_{RES} \cdot [\frac{1}{n} + \frac{\bar x^2}{s^2_x}]}]$$

e

$$[a_1 - t_{1- \alpha/2} \cdot \sqrt{s^2_{RES} \cdot \frac{1}{s^2_x}}, a_0 + t_{1- \alpha/2} \cdot \sqrt{s^2_{RES} \cdot \frac{1}{s^2_x}}]$$

dove $s^2_x = \sum (x_i - \bar x)^2$


Mentre per $\sigma^2$ un l'intervallo di confidenza è

$$[\frac{(n-2)S^2_{RES}}{q_{1-\alpha/2}}, \frac{(n-2)S^2_
{RES}}{q_{\alpha/2}}]$$

## Attendibilità modello lineare

Dopo aver stimato i parametri dobbiamo verificare l'attendibilità di questi. 

Si possono tranquillamente usare i test visti in precedenza (che permettodo di capire l'incorrelazione tra due caratteri), ma in caso di rifiuto può aver senso fare test più approfonditi sulla validità del modello lineare.

In caso in cui si posseggano dati accurati è utile utilizzare il test basato su 

$$T_n^\wedge = R_n \cdot \sqrt{\frac{n-2}{1-R_n^2}}$$

Mentre se i dati non sono accurati si usa il test basato su 

$$T_n^\sim = R_s \cdot \sqrt{\frac{n-2}{1-R_s^2}}$$

Un altro test che si può attuare è quello basato sull'**analisi della varianza**.

### Test analisi della varianza


Dato un campione di $\{(X_n, Y_n)\}$ e gli stimatori $A_0$, $A_1$ si considerano le seguenti statistiche

$$D_{TOT} = \sum (Y_i - \bar Y)^2 \text{ Devianza totale}$$
$$D_{SP} = \sum (Y_i^\wedge - \bar Y)^2 \text{ Devianza Spiegata}$$
$$D_{RES} = \sum (Y_i - Y^\wedge)^2 \text{ Devianza dei residui}$$

Si può dimostrare che $$D_{TOT} = D_{SP} + D_{RES}$$

Se non fosse per i residui allora la regressione spiegherebbe lo scostamento totale tra le osservazioni e la loro media. 

Il modello è adatto a descrivere variazioni del carattere $Y$ se queste sono causate dalla regressione ovvero se il rapporto $R^2 = \frac{D_{SP}}{D_{TOT}} = 1 - \frac{D_{RES}}{D_{TOT}}$ è prossimo a 1. 

Inoltre $\sqrt{R^2} = |r_{XY}|$ (cioè è il quadrato del coefficiente di correlazione) ma a differenza del coefficiente che ha senso solo per regressioni lineari semplici. $R^2$ è detto coefficiente di correlazione generalizzato

----

Il modello lineare è adatto a descrivere le variazioni del carattere $Y$ se la quantità $D_{RES}$ è molto inferiore a $D_{SP}$ cioè se $D_{SP}/D_{RES}$ è molto grande.

-----

Un test sull'attendibilità del modello lineare è quello basato sulla statistica

$$F^\sim = (n-2) \cdot \frac{D_{SP}}{D_{RES}} = D_{SP} / S^2_{RES}$$

che è distribuita come una F con (1, n-2) gradi di libertà.

Si può quindi effettuare un test per l'ipotesi nulla 

$$H_0: \alpha_1 = 0$$  

rifiutandola quando la realizzazione $f^\sim$ della statistica $F^\sim$ assume valori grandi ovvero quando $f^\sim > F_{1-\alpha}$ 

Si ricoda che per effettuare questo test c'è da verificare che 

* $E[E_i] = 0$
* $V[E_i] = \sigma^2$
* le variabili $E_i$ sono incorrelate tra loro
* le variabili $E_i$ sono normalmente distribuite

----

Il secondo test si basa sul fatto che, se ci fosse una relazione tra $X$ ed $Y$ allora $\alpha_1 \neq 0$ 

Se fosse infatti $\alpha_1 = 0$ allora le variazioni delle Y non dipenderebbero da quelle delle X. 

Si effettua quindi un test avente come ipotesi nulla 

$$H_0: \alpha_1 = 0$$

Si verificano i requisiti sui residui e si definisce $T_1$ come

$$T_1 = \frac{A_1 - \alpha_1}{\sqrt{S^2_{RES} \cdot \frac{1}{\sum(X_i - \bar X)^2}}}$$

In particolare visto che ipotizziamo $H_0$ soddisfatta allora

$$T^\sim = \frac{A_1}{\sqrt{S^2_{RES} \cdot \frac{1}{\sum(X_i - \bar X)^2}}}$$

che è distribuita come una t di Student con n-2 gradi di libertà.

Ad esempio per un test di ampiezza $\alpha$ rifiutiamo $H_0$ quando 

$$|t^\sim > t_{1 - \alpha/2}|$$

## Analisi dei residui

Abbiamo prima definito i residui 

$$E_i = Y_i^\wedge - Y_i$$ dove

$Y_i^\wedge = A_0 + A_1 \cdot X_i$

come la differenza tra i valori assunti e le sue stime.

Abbiamo anche fatto ipotesi sui residui, infatti essi:

* devono avere media nulla
* siano incorrelati
* abbiano identica varianza
* siano normalmente distribuiti

E' importante controllare la validità di tali ipotesi.

### P-value

E' impossibile sapere, senza ricalcolare l'intervallo, se, al cambiare del livello di significativita' $\alpha$, si accetta o si rifiuta l'ipotesi.

Un alternativa al calcolo dell'intervallo e' in questo caso il **p-value**. 

Il p-value e' definita come la probabilita' che la statistica valida sotto l'ipotesi nulla assuma valore nella region critica.

Il p-value misura la verosimiglianza del valore campionario ottenuto rispetto distribuzione statistica sotto l'ipotesi nulla.

Quanto **piu' piccolo e' il p-value** tanto piu' inverosimile e' che il risultato campionario della statistica provenga dalla distribuzionne della statistica valida sotto l'ipotesi nulla: con **tanta maggior sicurezza si rifiuta l'ipotesi nulla in favore della alternativa**.

Invece tanto e' piu' grande il p-value tanto e' maggiore la sicurezza che ci permette di non escliudere l'ipotesi nulla.

-----

Relativamente all'ipotesi di incorrelazione dei residui il metodo piu' utilizzato e' il test di Durbin-Watson.

$$D = \frac{\sum (E_i-E_{i-1})^2}{\sum E_i^2}$$

Tale statistica sotto l'ipotesi di incorrelazione ha una distribuzione particolare.

-----

Invece per l'ipotesi di normalita' si possono usare vari test, tra cui quello per la bonta' di adattamento.

Presentiamo quello di KS, del Chi-Quadro e di Shapiro Wilk.

In tutti i casi si controlla che i residui siano distribuiti con una media nulla e varianza $S^2_{RES} = \frac{1}{n-2} \sum (Y_i - Y_i^\wedge)^2$

**Test di KS**

E' basato su

$$K_n = sup |F_e(x) - F_n(x)| \cdot \sqrt{n}$$

Si rifiuta l'ipotesi di normalita' se $K_n$ e' grande. 

Richiede campioni di elevata numerosita'.

**Test di Shapiro Wilk**

Serve per campioni di bassa numerosita'.

$$W = \frac{[\sum w_i e_i]^2}{\sum (e_i - \bar e)^2}$$


La statistica W misura la rettilineita' del normal plot.

Si rifiuta se W e' piccolo.

-----

Per l'ipotesi di Omoschedasticita' (ovvero uguale varianza per ogni residuo) si hanno molte difficolta'.

Test veramente significativi per tale ipotesi possono essere realizzati se il numero di valori assumibili dal carattere X è limitato e se per ognuno di tali valori esistono numerose osservazioni.

Questa situazione purtroppo si presenta raramente nelle applicazioni e si è costretti a ricorrere a metodi grafici.

## Regressione lineare multipla

$$Y_i = \alpha_0 + \alpha_1 \cdot X_{1,i} + \ldots + \alpha_m \cdot X_{m,i} + E_i$$

Anche in questo caso i residui devono sottostare a determinate ipotesi (media a 0, varianza a $\sigma^2, incorrelati)$
